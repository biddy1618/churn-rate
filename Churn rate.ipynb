### Some note on number of features 'thumb-rule'
<a source='https://bioinformatics.oxfordjournals.org/content/21/8/1509.full'>This paper</a> suggests the following:

For uncorrelated features, the optimal feature size is $N-1$ (where $N$ is sample size)
As feature correlation increases, and the optimal feature size becomes proportional to $\sqrt{N}$ for highly correlated 
features.
Another (empirical) approach that could be taken, is to draw the learning curves for different sample sizes from the same dataset, and use that to predict classifier performance at different sample sizes. Here's the link to the paper.
***
What you've hit on here is the curse of dimensionality or the $p>>n$ problem (where $p$ is predictors and $n$ is observations). There have been many techniques developed over the years to solve this problem. You can use _AIC_ or _BIC_ to penalize models with more predictors. You can choose random sets of variables and asses their importance using _cross-validation_. You can use _ridge-regression_, the _lasso_, or _the elastic net_ for regularization. Or you can choose a technique, such as a _support vector machine_ or _random forest_ that deals well with a large number of predictors.

Honestly, the solution depends on the specific nature of the problem you are trying to solve.
***

Let's build some baseline classification models for our dataset. We will be dividing our dataset into train-test split with test ration _0.2_, we will be cross-validating our train set using 5-fold cross-validation scheme to get the metrics of our model, then we will get the quality metric of our model using test set. 

We will be using __Logistic regression__, and __Random forest__ classifiers. __Random Forest__ is kind of "art state" statistical method. It's fairly good for training even small samples. However, it fails when there are rare outcomes or rare predictors, as the algorithm is based on bootstrap sampling. In other words, if we have number of outliers, __Random Forest__ is prone to overfitting. Also it is less interpretable and needs some parameter tuning, but in our case, since we have really small data, it shouldn't take much time. __Logistic Regression__ in other hand, is more robust in terms of stability, handling out of range input data, and interpretability. It can better handle sparse data than __Random Forest__. It faster on big datasets, more interpretable as well. We will use both, and choose the best one. But most likely the __Random Forest__ will outperform __Logistic Regression__ since we have small non-sparse dataset (approximately ~20000 records with ~100 features), also this <a h='https://bmcbioinformatics.biomedcentral.com/articles/10.1186/s12859-018-2264-5'>link</a> confirms that.

We will use the __ROC-AUC__ metric for the quality measure of our models, since we have very disbalanced target variable. First, let's drop the duplicates, if we have some.

***
### Side note on models
* Both algorithms are really fast. There isn't much to distinguish them in terms of run-time.
* Logistic regression will work better if there's a single decision boundary, not necessarily parallel to the axis.
* Decision trees can be applied to situations where there's not just one underlying decision boundary, but many, and will work best if the class labels roughly lie in hyper-rectangular regions.
* Logistic regression is intrinsically simple, it has low variance and so is less prone to over-fitting. Decision trees can be scaled up to be very complex, are are more liable to over-fit. Pruning is applied to avoid this.

Source: <a href='https://www.quora.com/What-are-the-advantages-of-logistic-regression-over-decision-trees-Are-there-any-cases-where-its-better-to-use-logistic-regression-instead-of-decision-trees'>What are the advantages of logistic regression over decision trees? Are there any cases where it's better to use logistic regression instead of decision trees?</a>

Setting up the train-test splits, random seed, and target variable

```python
seed = 17
train_df, test_df = train_test_split(df, test_size=0.2, random_state=seed)
X_train, X_test = train_df.drop(columns=targ).values, test_df.drop(columns=targ).values
y_train, y_test = train_df[targ].values, test_df[targ].values
```

Training the logistic regression on the training dataset, without any feature engineering and feature selection.

```python
lg = LogisticRegression(solver='liblinear', random_state=seed)
def_score_lg = cross_val_score(lg, X_train, y_train, cv=5, scoring='roc_auc', n_jobs=-1).mean()
print('The cross-validation score on train data using LG is %.6f'%def_score_lg)
lg.fit(X_train, y_train)
print('The roc-auc score for our test set using LG is %.6f'%roc_auc_score(y_test, lg.predict_proba(X_test)[:,1]))
```

```python
feature_importances = pd.DataFrame(np.abs(lg.coef_.flatten()*np.std(X_train, 0).shape),
                                   index = train_df.drop(columns=targ).columns,
                                   columns=['importance']).sort_values('importance', ascending=False)
feature_importances.T
```

We got very low ROC-AUC score. We will do some feature engineering and selection next to improve that value. We want to keep the Logistic Regression to have model to compare with the Random Forest.

```python
rf = RandomForestClassifier(n_estimators=10, criterion='entropy', random_state=seed, n_jobs=-1)
def_score_rf = cross_val_score(rf, X_train, y_train, cv=5, scoring='roc_auc', n_jobs=-1).mean()
print('The cross-validation score on train data using RF is %.6f'% def_score_rf)
rf.fit(X_train, y_train)
print('The roc-auc score for our test set using RF is %.6f'%roc_auc_score(y_test, rf.predict_proba(X_test)[:,1]))
```

```python
feature_importances = pd.DataFrame(rf.feature_importances_,
                                   index = train_df.drop(columns=targ).columns,
                                   columns=['importance']).sort_values('importance', ascending=False)
feature_importances.T
```

We got relatively good results with Random Forest. As has been <a href='https://www.quora.com/Is-feature-engineering-relevant-at-all-for-Random-Forests-How-are-centering-scaling-and-Box-Cox-transformations-relevant-in-this-context'>said</a>, Random Forest is invariant to monotone (like linear) transformation of data. Thus we will try to beat this score with Logistic Regression by applying some feature engineering and feature selection.

## Feature engineering and selection

The feature importance for `month` by LG: 0.042702	
The feature importance for `month` by RF: 0.007993	

Let's test our hyphotesis that `month` feature might be irrelavant feature.

```python
train_df, test_df = train_test_split(df.drop(columns='month'), test_size=0.2, random_state=seed)
X_train, X_test = train_df.drop(columns=targ).values, test_df.drop(columns=targ).values
y_train, y_test = train_df[targ].values, test_df[targ].values

print('The baseline1 score on LG is                %.6f'%def_score_lg)
print('The CV score on LG without month feature is %.6f'% cross_val_score(lg, X_train, y_train, cv=5,
                                                                          scoring='roc_auc', n_jobs=-1,).mean())
```

Seems like `month` feature does improve our model slightly, but from the logical context it should not add any value to our model (especially in this numeric type). Let's convert this feature into dummy variable and test it again. 

```python
df = pd.get_dummies(df, columns=['month'])
train_df, test_df = train_test_split(df, test_size=0.2, random_state=seed)
X_train, X_test = train_df.drop(columns=targ).values, test_df.drop(columns=targ).values
y_train, y_test = train_df[targ].values, test_df[targ].values

print('The baseline1 score on LG is                    %.6f'%def_score_lg)
print('The CV score on LG with month dummy encoding is %.6f'% cross_val_score(lg, X_train, y_train, 
                                                                              cv=5, scoring='roc_auc', 
                                                                              n_jobs=-1,).mean())
```

This looks better than it was. Let's continue. We should scale the features, since logistic regression assumes normalized and standardized data as input.

```python
num_f = num_f.drop(['month'])
df[num_f].describe().T
```

We should standardize all numeric features except for `totalcallduration`, since this feature seems to have a skewed distribution, as we can observe below. 

```python
fig, axes = plt.subplots(2,2, figsize=(15,10))
fig.suptitle('totalcallduration feature transformation')
axes = axes.flatten()
tmp0 = df[df[targ]==0]['totalcallduration']
tmp1 = df[df[targ]==1]['totalcallduration']

sns.distplot(a=tmp0, hist=True, kde=False, color='blue', label='Not churned', ax=axes[0])
sns.distplot(a=tmp1, hist=True, kde=False, color='green', label='Churned',ax=axes[0])
axes[0].legend()
axes[0].set_title('Original distribution')

sns.distplot(a=np.sqrt(tmp0), hist=True, kde=False, color='blue', label='Not churned', ax=axes[1])
sns.distplot(a=np.sqrt(tmp1), hist=True, kde=False, color='green', label='Churned',ax=axes[1])
axes[1].legend()
axes[1].set_title('Square root transformation distribution')

sns.distplot(a=np.cbrt(tmp0), hist=True, kde=False, color='blue', label='Not churned', ax=axes[2])
sns.distplot(a=np.cbrt(tmp1), hist=True, kde=False, color='green', label='Churned',ax=axes[2])
axes[2].legend()
axes[2].set_title('Cube root transformation distribution')

sns.distplot(a=np.log(tmp0+1), hist=True, kde=False, color='blue', label='Not churned', ax=axes[3])
sns.distplot(a=np.log(tmp1+1), hist=True, kde=False, color='green', label='Churned',ax=axes[3])
axes[3].legend()
axes[3].set_title('Natural logarithm transformation distribution')

plt.tight_layout(pad=5)
plt.show()
```

Seems like _square root_ is the most reasable way to transform this feature, since it is the most similar to normal distribution. Let's apply that transformation to `totalcallduration` feature, and standize all numeric features.

```python
df['totalcallduration'] = np.sqrt(df['totalcallduration'])

ss = StandardScaler()
df[num_f] = ss.fit_transform(df[num_f])

df[num_f].describe().T
```

Let's test our data with new standardized features on LG. Also, for interest, let's test the new data with RF.

```python
train_df, test_df = train_test_split(df, test_size=0.2, random_state=seed)
X_train, X_test = train_df.drop(columns=targ).values, test_df.drop(columns=targ).values
y_train, y_test = train_df[targ].values, test_df[targ].values

print('The baseline1 score on LG is                     %.6f'%def_score_lg)
print('The CV score on LG with standardized features is %.6f'% cross_val_score(lg, X_train, y_train, 
                                                                               cv=5, scoring='roc_auc', 
                                                                               n_jobs=-1).mean())
lg.fit(X_train, y_train)
print('The cross-validation score on test using RF is   %.6f'% roc_auc_score(y_test, 
                                                                           lg.predict_proba(X_test)[:,1]))
```

```python
print('The baseline1 cross-validation score using RF is %.6f'% def_score_rf)
print('The CV score on standardized data using RF is    %.6f'% cross_val_score(rf, X_train, y_train, 
                                                                               cv=5, scoring='roc_auc', 
                                                                               n_jobs=-1).mean())
rf.fit(X_train, y_train)
print('The CV score on test using RF is                 %.6f'% roc_auc_score(y_test, 
                                                                             rf.predict_proba(X_test)[:,1]))
```

The standardized data increased the quality metric for Logistic regression, but slightly decreased the cross-validation score on Random forest (but in test data, it showed slight improvement). So, let's keep our dataset in this shape.

## Optimizing hyperparameters

We will optimize the hyperparameters of models, and see the important features. 

```python
C = np.logspace(-2,0,10)

lg = LogisticRegression(solver='liblinear')
gsCV = GridSearchCV(lg, param_grid={'C': C, 'penalty':['l1', 'l2']}, 
                    scoring='roc_auc', cv=5, verbose=1, n_jobs=-1)

gsCV.fit(X_train, y_train)
print('The best parameter for Logistic regression is:', gsCV.best_params_)
print('The CV score on test data for optimzed parameters:', gsCV.best_score_)
```

```python
print('The score on test data is:', roc_auc_score(y_test, gsCV.predict_proba(X_test)[:,1]))
print('Time for LG parameter search in seconds: ', 21.5)
```

Feature importance based on Logistic Regression

```python
feature_importances_lr = pd.DataFrame(np.abs(gsCV.best_estimator_.coef_.flatten()),
                                   index = train_df.drop(columns=targ).columns,
                                   columns=['importance']).sort_values('importance', ascending=False)
feature_importances_lr.T
```

Grid search for Random Forest model (this might take a while to run, since Random Forest has quite lots of hyperparameters).

```python
from sklearn.model_selection import RandomizedSearchCV

n_estimators = [int(x) for x in np.linspace(start = 10, stop = 200, num = 10)]
max_features = ['auto', 'sqrt']
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)
min_samples_split = [2, 5, 10]
min_samples_leaf = [1, 2, 4]
bootstrap = [True, False]
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}


rf = RandomForestClassifier()
rsCV = RandomizedSearchCV(estimator=rf, param_distributions=random_grid, n_iter=100, 
                          scoring='roc_auc', cv=5, verbose=2, random_state=seed, n_jobs=-1)
rsCV.fit(X_train, y_train)
print('The best parameter for Logistic regression is:', rsCV.best_params_)
print('The CV score on test data for optimzed parameters:', rsCV.best_score_)
```

```python
print('The score on test data is:', roc_auc_score(y_test, rsCV.predict_proba(X_test)[:,1]))
print('Time for RF parameter search in minutes:', 13.1)
```

Feature importance based on Random Forest

```python
feature_importances_rf = pd.DataFrame(np.abs(rsCV.best_estimator_.feature_importances_),
                                   index = train_df.drop(columns=targ).columns,
                                   columns=['importance']).sort_values('importance', ascending=False)
feature_importances_rf.T
```
